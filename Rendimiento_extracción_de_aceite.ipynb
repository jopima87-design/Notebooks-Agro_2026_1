# **Control de Rendimiento en Extracción de Aceite** #
Este ejercicio es ideal para introducir float, int y la concatenación de strings.

# **Planteamiento:** #
Una planta procesadora de palma de aceite necesita calcular el Índice de Rendimiento de Extracción (IRE). El operario debe ingresar el peso total de los racimos de fruta fresca (RFF) procesados en kilogramos y la cantidad de aceite crudo obtenido en litros.

# **Requerimientos:** #
1. Solicitar el nombre del lote de producción (Tipo string).
2. Solicitar el peso de los racimos en kg (Tipo float). (pr)
3. Solicitar los litros de aceite obtenidos (Tipo float).(lt)
4. Calcular el rendimiento sabiendo que la densidad aproximada del aceite es de 0.92 kg/L. (r)
.

$$ pa (kg)=lt * 0.92 $$
$$ r =(pa/pr) * 100 $$
**Prueba del programa:**

Si se ingresan 1000 kg de racimos y se obtienen 200 litros de aceite, el programa debe informar que el rendimiento del lote es del 18.4%.

```python
# ENTRADA: Solicitar el nombre del lote de producción (Tipo string)
# Solicitar el peso de los racimos en kg (Tipo float). (pr)
# Solicitar los litros de aceite obtenidos (Tipo float).(lt)
```
Estas líneas son comentarios que describen las tres entradas que el programa solicitará al usuario. Son notas para entender el propósito de las siguientes líneas de código.

```python
NOMBRE = input("Ingrese el nombre del lote: ")
```
Esta línea pide al usuario que ingrese el nombre del lote de producción. El texto "Ingrese el nombre del lote: " se muestra en la consola, y lo que el usuario escribe se guarda como una cadena de texto en la variable `NOMBRE`.

```python
pr = float(input("Ingrese el peso de los racimos en kg: "))
```
Esta línea pide al usuario el peso de los racimos en kilogramos. Similar a la anterior, muestra el mensaje, pero la función `float()` convierte la entrada del usuario (que siempre es una cadena) a un número decimal (tipo float) antes de guardarlo en la variable `pr`.

```python
lt = float(input("Ingrese la cantidad de aceite crudo obtenido en litros: "))
```
Esta línea pide al usuario la cantidad de aceite crudo obtenido en litros. De nuevo, el texto se muestra, y la entrada del usuario se convierte a un número decimal (`float`) y se almacena en la variable `lt`.

```python
# PROCESO: Calcular el peso del aceite (pa) y el rendimiento (r)
```
Este es un comentario que indica que las siguientes líneas de código realizarán los cálculos necesarios para obtener el peso del aceite y el rendimiento.

```python
pa = lt * 0.92
```
Esta línea calcula el peso del aceite en kilogramos (`pa`). Multiplica los litros de aceite obtenidos (`lt`) por la densidad aproximada del aceite, que es 0.92 kg/L. El resultado se guarda en la variable `pa`.

```python
r = (pa/pr) * 100
```
Esta línea calcula el Índice de Rendimiento de Extracción (IRE) en porcentaje (`r`). Divide el peso del aceite (`pa`) entre el peso de los racimos (`pr`) y luego multiplica el resultado por 100 para expresarlo como un porcentaje.

```python
# SALIDA: informar el rendimiento del lote
```
Este es un comentario que indica que la siguiente línea de código mostrará el resultado al usuario.

```python
print("El rendimiento del lote", NOMBRE, "es del", r, "%")
```
Esta línea imprime el resultado final en la consola. Utiliza una serie de cadenas de texto y las variables `NOMBRE` y `r` para construir un mensaje claro que informa al usuario sobre el rendimiento calculado para el lote específico.
```

````python
# ENTRADA: Solicitar el nombre del lote de producción (Tipo string)
# Solicitar el peso de los racimos en kg (Tipo float). (pr)
# Solicitar los litros de aceite obtenidos (Tipo float).(lt)
NOMBRE = input("Ingrese el nombre del lote: ")
pr = float(input("Ingrese el peso de los racimos en kg: "))
lt = float(input("Ingrese la cantidad de aceite crudo obtenido en litros: "))

# PROCESO: Calcular el peso del aceite (pa) y el rendimiento (r)
pa = lt * 0.92
r = (pa/pr) * 100

# SALIDA: informar el rendimiento del lote
print("El rendimiento del lote", NOMBRE, "es del", r, "%")
````
