### 1. Cálculo del Índice de Madurez

En la agroindustria, la relación entre Sólidos Solubles Totales (Brix) y la acidez es clave para determinar la cosecha.

```
# Datos de la muestra
grados_brix = 12.5
acidez_titulable = 0.8

# Cálculo del índice
indice_madurez = grados_brix / acidez_titulable

print(f"El índice de madurez es: {indice_madurez}")

if indice_madurez > 15:
    print("Estado: Óptimo para procesamiento.")
else:
    print("Estado: Aún le falta maduración.")
```

### 2. Balance de Masa (Mezclado Simple)

Imagina que quieres estandarizar el contenido de grasa en una mezcla láctea.

```
# Queremos saber cuánta leche A (3% grasa) y leche B (12% grasa)
# necesitamos para obtener 100kg de mezcla al 5% de grasa.

def balance_grasa(total_deseado, porcentaje_final, grasa_a, grasa_b):
    # Ecuación simple de mezcla
    masa_b = (total_deseado * (porcentaje_final - grasa_a)) / (grasa_b - grasa_a)
    masa_a = total_deseado - masa_b
    return masa_a, masa_b

leche_a, leche_b = balance_grasa(100, 5, 3, 12)

print(f"Mezclar {leche_a:.2f} kg de Leche A con {leche_b:.2f} kg de Leche B.")
```

### 3. Estimación de Pérdida de Peso en Almacenamiento

Un cálculo rápido para monitorear la merma por transpiración en productos frescos.

```
peso_inicial = 500.0  # kg
peso_final = 485.5    # kg

# Cálculo del porcentaje de merma
merma = ((peso_inicial - peso_final) / peso_inicial) * 100

print(f"La pérdida de peso fue del {merma}%")

if merma > 3.0:
    print("Alerta: Revisar humedad relativa en cámara frigorífica.")
```
